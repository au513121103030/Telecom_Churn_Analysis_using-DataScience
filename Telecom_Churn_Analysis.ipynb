```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('telecom_churn.csv')
```

```python
# Display basic information about the dataset
print(df.info())
```

```python
# Display the first few rows of the dataset
print(df.head())
```

```python
# Summary statistics
print(df.describe())
```

```python
# Countplot for gender
plt.figure(figsize=(8, 6))
sns.countplot(x='gender', hue='churn', data=df)
plt.title('Churn Count by Gender')
plt.show()
```

```python
# Boxplot for age and churn
plt.figure(figsize=(8, 6))
sns.boxplot(x='churn', y='age', data=df)
plt.title('Age Distribution by Churn')
plt.show()
```

```python
# Barplot for state and churn
plt.figure(figsize=(12, 8))
sns.countplot(x='state', hue='churn', data=df)
plt.title('Churn Count by State')
plt.xticks(rotation=45)
plt.show()
```

```python
# Barplot for telecom partner and churn
plt.figure(figsize=(12, 8))
sns.countplot(x='telecom_partner', hue='churn', data=df)
plt.title('Churn Count by Telecom Partner')
plt.xticks(rotation=45)
plt.show()
```
